```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Define X and y_true here
# For example:
X = np.array([[1], [2], [3], [4], [5]])  # Example input data
y_true = np.array([2, 4, 5, 4, 5])  # Example true output data

# Add noise to the data
noise = np.random.normal(0, 5, X.shape[0])  # Adjust noise level to see impact
y = y_true + noise  # Noisy observations
```

```python
# Fit a simple linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Plot the data
plt.scatter(X, y, label='Noisy Data')
plt.plot(X, y_true, label='True Relationship', linestyle='dashed', color='green')
plt.plot(X, y_pred, label='Regression Line', color='red')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.show()

```

```python
# Print model parameters
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
print(f'Mean Squared Error: {mean_squared_error(y, y_pred)}')
```
