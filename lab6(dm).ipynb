```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
#example data
data={'category':['A','B','C','A','B']}
df=pd.DataFrame(data)
df
```

```python
#label encoding
label_encoder=LabelEncoder()
df['Label_Encoder']=label_encoder.fit_transform(df['category'])
df
```

```python
#one hot encoding
one_hot_encoder=pd.get_dummies(df['category'],prefix='category')
df=pd.concat([df,one_hot_encoder],axis=1)
df
```

```python
# Import necessary libraries
import pandas as pd
from datetime import datetime

# Example data
data = {'Name': ['Alice', 'Bob'], 'Date_of_Birth': ['1990-05-14', '1985-08-23']}
df = pd.DataFrame(data)

# Convert 'Date_of_Birth' to datetime
df['Date_of_Birth'] = pd.to_datetime(df['Date_of_Birth'])

# Calculate age
current_year = datetime.now().year
df['Age'] = current_year - df['Date_of_Birth'].dt.year
df
```

```python
#example data
data={'category':['A','B','C','A','B'],'value':[10,20,10,30,40]}
df=pd.DataFrame(data)
df
```
